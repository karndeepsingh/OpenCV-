### Contour can be helpful for object detection or object recongintion or shape analysis. So, for finding contours we use binary images.

### Before finding the contour we are going to apply threshold and apply canny edge detection to find the contour on the image

```python
import cv2
import numpy as np
```

```python
img = cv2.imread("C:\\Users\\KARNDEEP SINGH\\Desktop\\opencv-master\\opencv-master\\samples\\data\\opencv-logo.png")
imggray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(imggray, 127,255, 0)
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)
print("Number of contours:" + str(len(contours)))
print(contours[0])

cv2.drawContours(img, contours, -1, (0,255,0), 3) #if we give -1 we can get contours over the whole images and for specific contours we need to pass index i.e for 0th index we pass 0  
cv2.imshow("opencv", img)
cv2.imshow("imageGray", imggray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
