# For Smoothing or Blurring the Images Linear kind of Filters used because they are each to achieve and they are relatively fast


 There are various type of filters available such as Gaussian, Homogeneous, Median, Bilateral filters etc...

1. HOMOGENOUS FILTER: It is the most simple filter each output pixels is the mean of the kernel neighbours. And in homegeous filter each pixel contribute with equal weights therefore it is called homogeneous filter.

In one dimensional signals images images can also be filtered with low pass filters(LPF) and High PASS FILTERs(HPF)

LPF - helps in REMOVING the noises and blureing the images

HPF -helps in finding the edges in the images

2. Gaussian filter: It is nothing but using different weight along different axis i.e x and y so the pixel located on center of filter has higher weight and on the side of image has lower weight.


3. Median Filter: It is the filter which replaces each pixel's value with median of its neighbouring pixel. This method is great when dealing with salt and pepper noise.

4. Bilateral Filter: The above filter method smoothens out the noise and edges. but when we need to preserve the edges of the images we need to bilateral filters.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("C:\\Users\\KARNDEEP SINGH\\Desktop\\opencv-master\\opencv-master\\samples\\data\\opencv-logo.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.uint8)/25 #it is as per formula of kernal that kernel = 1/(kernel width, kernel height)*[width x height matrix of ones]
dst = cv2.filter2D(img, -1,kernel) #convoluting the images and filter
blur = cv2.blur(img, (5,5)) #it is just averaging of the filters
gaussian_blur = cv2.GaussianBlur(img, (5,5), 0)

images = [img, dst, blur, gaussian_blur]
titles = ["images", "2D Convolution", "Blur", "gaussian blur"]

plt.figure(figsize = (10,10))
for i in range(4):
    plt.subplot(1,4, i+1), plt.imshow(images[i])
    plt.title(titles[i])
```
