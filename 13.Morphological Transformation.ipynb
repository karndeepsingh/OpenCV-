# Morphological Transformation are performed on Binary Images therefore we are transforming Colorful images into Grayscale images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:\\Users\\KARNDEEP SINGH\\Desktop\\opencv-master\\opencv-master\\samples\\data\\smarties.png", 0)
# img = cv2.imread("C:\\Users\\KARNDEEP SINGH\\Desktop\\opencv-master\\opencv-master\\samples\\data\\j.png", cv2.IMREAD_GRAYSCALE)
_, mask = cv2.threshold(img, 220,255, cv2.THRESH_BINARY_INV)

kernel = np.ones((3,3), np.uint8)

#types of morphological operations
dilation = cv2.dilate(mask, kernel, iterations = 2) #it is applied to remove some dots on the balls and as iteration increases dots are gone but balls edges get into eachother.
erosion = cv2.erode(mask, kernel, iterations = 2) #it is applied to erode the edges of the balls.
opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN , kernel, iterations = 2)#opening is an erosion followed by dilation
closing = cv2.morphologyEx(mask,cv2.MORPH_CLOSE , kernel, iterations = 2)#closing is an dilation followed by erosion
morphy_gradient = cv2.morphologyEx(mask,cv2.MORPH_GRADIENT , kernel, iterations = 2)#it is the difference between opening and closing morphology
morphy_tophat= cv2.morphologyEx(mask,cv2.MORPH_TOPHAT, kernel, iterations = 2) #it is the difference between image and opening


images =[img, mask, dilation, erosion, opening, closing, morphy_gradient, morphy_tophat]
titles = ["original", "Mask", "Dilation","Erosion", "opening", "closing","morphy_gradient","morphy_tophat"]

plt.figure(figsize =(10,5))
for i in range(8):
    plt.subplot(2,4, i+1), plt.imshow(images[i], "gray")
    plt.title(titles[i])
```
