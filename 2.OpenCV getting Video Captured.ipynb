```python
import cv2
```

## Capturing the Video from camera

```python
cap = cv2.VideoCapture(0); # 0 indicates the 0th index camera i.e we have web cam so it will use that. if we have more than 1 camera and activate all those then we can use repective index values like 1,2 so on.
#we can also give "Filename.format" inside the VideoCapture function as argument.
while (True):
    ret, frame = cap.read() #reading the frame captured by video and reutrn True and False i.e if Frame is available then it is True else False
    
    #we can also convert the image colour to Gray scale
#     frame = cv2.cvtColor(frame, cv2.COLOR_BG2GRAY)
    cv2.imshow("video", frame)
    
    if cv2.waitKey(1) == ord("q"):
        break

cap.release() #release the cap variable so that all resoures are realsed
cv2.destroyAllWindows() #destroy all variable
```

## Displaying Properties of the video being captured


```python
cap = cv2.VideoCapture(0)
print(cap.isOpened())
while (cap.isOpened()): #cap.isOpened() will return Boolean Values SO, it Return True When index passed in source is True otherwise False
    ret, frame = cap.read()
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) #cap.get command is use to print the repective properties of the frame
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        
    cv2.imshow("video",frame)
        
    if cv2.waitKey(1)== ord("q"):
            break
        
cap.release()
cv2.destroyAllWindows()
```

## Writing the Video or Saving The Video

```python
cap = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*"XVID")
out = cv2.VideoWriter("output.avi", fourcc, 20.0, (640,480))

while (cap.isOpened()):
    ret, frame = cap.read()
    if ret ==True:
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        
        out.write(frame)
        
        #gray = cV2.cvtColor(frame, COLOR_BG2GRAY)
        
        cv2.imshow("video", frame)
        
        if cv2.waitKey(1) == ord("q"):
            break
    else:
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
